```python
import numpy as np
import pandas as pd
from scipy.stats import binom
import matplotlib.pyplot as plt
%matplotlib inline


plt.subplots(1,1)
```

```python
fig, ax  = plt.subplots(1,1)
x = [0,1,2,3,4,5,6]
n = 6
p = 0.5
rv = binom(n,p) #Generates a random variable with binomial distribution

ax.vlines(x,0, rv.pmf(x)) #vlines = vertical lines used to plot vertical
                          #lines
plt.show()
```

```python
fig, ax  = plt.subplots(1,1)
x = [0,1,2,3,4,5,6]
n = 6
p = 0.5
rv = binom(n,p) #Generates a random variable with binomial distribution

ax.vlines(x,0, rv.pmf(x), colors = "k", linestyle ='--', lw=1,
         label= "Probability")
ax.legend(loc = "best", frameon = "False")
plt.show()
```

```python
#Z score:
# A z score is the score tha exresses the value of distribution in std dev
# w.r.t to mean.
# z = (x-mu)/sigma

#Example: A classroom has 60 students in it and they have just got math
#examp score. Simulat this with normal distribution

classscore = np.random.normal(50,10,60).round() # n =60,mu=mean =50,sigma=stddev=10
plt.hist(classscore,30, normed = True) # of breaks = 30
plt.show()
```

```python
#score of each student can be converted into a zscore
stats.zscore(classscore)
```

```python
print("Score of a student:",classscore[34])
print("z-score of the same student", stats.zscore(classscore)[34])
```

```python
#What is the probabilty of a student getting 64 or above.
prob = 1 - stats.norm.cdf(1.693)
#cdf funciton give the probability of getting values upto the z-score of
#1.693. So 1-that value will give the probability of scoring above that val

prob
# i.e there 4.5% chance of getting a score of 64 or above.
```

```python
#How many students made it to the top 20% of the class.

#At what z-score in the normal distribution is 80% of pop in one side and 
# 20% population on one side. use ppf fucntion in scipy

print("80% marging in normal distribution is:", stats.norm.ppf(0.8))
# Hence Z-score of 0.8416 or above is needed to be in top 20% of the class

#z = x-mu/sigma => z*sigma+mu = x
val = 0.8416*classscore.std()+classscore.mean()
print("Least score needed to be in top 20% of class:", val)


```

```python
# A p-value
# p-value is the probability of rejecting null hypothesis when the hypothesis
# is proven true

#Null Hypothesis: Common for students to score 68 marks.

#1.Calculcate zscore
zscore = (68-classscore.mean())/classscore.std()

#Calculate probability of getting this zscore and above
prob = 1- stats.norm.cdf(zscore)

print("The probability of having score of 68 and above:", prob)
#Probability is only 1.6% i.e p value is very small
# Hence null hypothesis can be rejected
```
