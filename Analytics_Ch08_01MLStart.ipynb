```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from sklearn import datasets
%matplotlib inline

iris = datasets.load_iris()
print(iris.data[0:5,:],"\n") #iris data set value
print(iris.target) #Targetted label values for iris
print(iris.target_names)
```

```python
#Building a scatter plot
x = iris.data[:,0] #X-Axis = sepal length
y = iris.data[:,1] # Y-Axis = sepal width
species = iris.target #lables for speciies

#creating axis limits for plottting
x_min = x.min()-0.5 #subtracting 0.5 for easy visualization of end points
x_max = x.max()+0.5
y_min,y_max = y.min()-0.5, y.max()+0.5

#create scatter plot
plt.figure()
plt.title("Iris dataset- Classification by Sepal Sizes")
plt.scatter(x,y, c = species) #plot x and y, class - species from iris label
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.xlim(x_min, x_max) #X graphs limits
plt.ylim(y_min, y_max) #y graphs limit
#plt.xticks(())
#plt.yticks(())
```

```python
#Building a scatter plot for petals
x = iris.data[:,2] #Petal length
y = iris.data[:,3] #petal width
species = iris.target

#creating axis limits for plotting
x_min, x_max = x.min()-0.5, x.max()+0.5
y_min, y_max = y.min()-0.5, y.max()+0.5

plt.figure()
plt.title("Iris Dataset - Classification by Petal Sizes")
plt.scatter(x,y,species) #scatter plot: x, y = axis, species = color/value
plt.xlabel("Petal length")
plt.ylabel("Petal Width")
plt.xlim(x.min(),x.max())
plt.ylim(y.min(), y.max())
print(x_min, x_max, y_min, y_max)
```

```python
#Applying PCA to the data

#from sklearn.decomposition import PCA 
#x_reduced = PCA(n_components =3).fit_transform(iris.data)
    #calling PCA with 3 principal components.
    #fit transform():- pass the data with dimensions to it.

#PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA

iris = datasets.load_iris()
species = iris.target #Species
x_reduced = PCA(n_components = 3).fit_transform(iris.data)

#Scatter plot
fig = plt.figure()
ax = Axes3D(fig)
ax.set_title("Iris Dataset by PCA", size=14)
ax.scatter(x_reduced[:,0],x_reduced[:,1],x_reduced[:,2], c = species)
ax.set_xlabel("First Eigenvector")
ax.set_ylabel("Second Eigenvector")
ax.set_zlabel("Third Eigenvector")
```

```python
#KNN Nearest neighbor

import numpy as np
from sklearn import datasets
np.random.seed(0)
iris = datasets.load_iris()
x = iris.data
y = iris.target
i = np.random.permutation(len(iris.data))
    #Generate a list that is random permutation of numbers from 0 to
    # length of iris data (ie.# of rows)
len(i[:-10]) # 0 to -10elements in i i.e all elements in i except last 10

#Leaving 10 elements for testing and selecting everything else for training
x_train = x[i[0:-10]]
y_train = y[i[0:-10]]

#10 random elements for testing
x_test =x[i[-10:]]
y_test =y[i[-10:]]

from sklearn.neighbors import KNeighborsClassifier #Import KNN module

knn = KNeighborsClassifier() #Assign KNeighborClassifier to knn
knn.fit(x_train,y_train)     #classify the given data,label using knn

#knn model is now available. now predict
print(knn.predict(x_test))

#Actual answer 
print(y_test)
```

```python
print(knn.predict(x_test) == y_test)
# 1 false in 10 ie 10% error rate.
```
