```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv("http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv",
                  index_col = 0)
```

```python
data.head()
```

```python


fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize = (16,8))
ax1.scatter(data["TV"], data["Sales"])
ax1.set_xlabel = "TV Advertising"
ax1.set_ylabel = " Sales "

ax2.scatter(data["Radio"], data["Sales"])
ax2.set_xlabel  =  "Radio Advertising"
ax2.set_ylabel  = "Sales"

ax3.scatter(data["Radio"], data["Sales"])
ax3.set_xlabel  =  " Newspaper"
ax3.set_ylabel  = "Sales"
```

```python
from sklearn.linear_model import LinearRegression

tv = data["TV"].reshape(-1,1)
radio = data["Radio"].reshape(-1,1)
np = data["Newspaper"].reshape(-1,1)
sales = data["Sales"].reshape(-1,1)

#Linear Regression
lm = LinearRegression()
lm_fit = lm.fit(tv, sales)

print("Slope: ", lm.coef_)
print("Intercept: ", lm.intercept_)
```

```python
fig,ax1 = plt.subplots(1,1)
ax1.scatter(data["TV"], data["Sales"])
ax1.set_xlabel = "TV Advertising"
ax1.set_ylabel = " Sales "

ax1.plot(tv,lm.predict(tv), color = "r", lw = 3.0)
```

```python
#Multi linear
cols = ["TV","Radio", "Newspaper"]
X = data[cols]
Y = data["Sales"]

mlm = LinearRegression()
mlm_fit = mlm.fit(X,Y)
print("Slope: ", mlm.coef_)
print("Intercept: ", mlm.intercept_)
```

```python
import statsmodels.formula.api as smf
mlm = smf.ols(formula ="Sales~TV", data = data)
mlm_fit = mlm.fit()
print("Parameters of the linear fit\n",mlm_fit.params)
```

```python
#For creatng the fit, need a datafram
x = pd.DataFrame({"TV":list(data["TV"])})
y= mlm_fit.predict(x)

fig,ax1 = plt.subplots(1,1)
ax1.plot(x,y, color = "r", lw =3.0)
ax1.scatter(data["TV"], data["Sales"])
ax1.set_xlabel = "TV Advertising"
ax1.set_ylabel = " Sales "

#mlm = smf.ols(formula ="Sales~TV+Radio+Newspaper", data = data)
#mlm_fit = mlm.fit()
#mlm_fit.predict()
```
