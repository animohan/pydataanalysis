```python
#Many properties for the plot
import numpy as np
import pandas as pd
import pandas.tools.rplot as rplot
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot([1,2,3,4],[1,4,9,16])
plt.show
```

```python
plt.plot([1,2,3,4],[1,4,9,16], linewidth = 4.0, color = 'g')
```

```python
#Plot function returns the list of line objects. Using the line setter
#methods can define the property that needs to be set
line, = plt.plot([1,2,3,4],[1,4,9,16])
line.set_linestyle('--') #setting dashed line
plt.show()
```

```python
#Using the setp() command
line, = plt.plot([1,2,3,4],[1,4,9,16])
plt.setp(line, color = "r", linewidth = 2.0) #setting line width & color
plt.show()
```

```python
#Creating multiple plots
p1 = np.arange(0.0, 30.0, 0.1)
plt.subplot(211) #Subplot with 2 rows and 1 column and 1 figure
plt.plot(p1, np.sin(p1)/p1, 'b--') #blue dashed line
plt.subplot(212)
plt.plot(p1, np.cos(p1),'r--')
```

```python
plt.subplot(221)
plt.plot(p1, np.sin(p1),'g--')
plt.subplot(222) #Subplot with 2 rows and 2 column and 2 figure
plt.plot(p1, np.sin(p1)/p1, 'b--') #blue dashed line
plt.subplot(212)
plt.plot(p1, np.cos(p1),'r--')
```
