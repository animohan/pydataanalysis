```python
import numpy as np
import pandas as pd
from scipy.stats import binom
import matplotlib.pyplot as plt
%matplotlib inline


plt.subplots(1,1)
```

```python
fig, ax  = plt.subplots(1,1)
x = [0,1,2,3,4,5,6]
n = 6
p = 0.5
rv = binom(n,p) #Generates a random variable with binomial distribution

ax.vlines(x,0, rv.pmf(x)) #vlines = vertical lines used to plot vertical
                          #lines
plt.show()
```

```python
#Z score:
# A z score is the score tha exresses the value of distribution in std dev
# w.r.t to mean.
# z = (x-mu)/sigma

#Example: A classroom has 60 students in it and they have just got math
#examp score. Simulat this with normal distribution

classscore = np.random.normal(50,10,60).round() 
            # n =60,mu=mean =50,sigma=stddev=10
plt.hist(classscore,30, normed = True) # of bins = 30
plt.show()
```

```python
#score of each student can be converted into a zscore
stats.zscore(classscore)
```

```python
print("Score of a student:",classscore[34])
print("z-score of the same student", stats.zscore(classscore)[34])
```

```python
#What is the probabilty of a student getting 64 or above.
prob = 1 - stats.norm.cdf(1.693)
#cdf funciton give the probability of getting values upto the z-score of
#1.693. So 1-that value will give the probability of scoring above that val

prob
# i.e there 4.5% chance of getting a score of 64 or above.
```

```python
#How many students made it to the top 20% of the class.

#At what z-score in the normal distribution is 80% of pop in one side and 
# 20% population on one side. use ppf fucntion in scipy

print("80% marging in normal distribution is:", stats.norm.ppf(0.8))
# Hence Z-score of 0.8416 or above is needed to be in top 20% of the class

#z = x-mu/sigma => z*sigma+mu = x
val = 0.8416*classscore.std()+classscore.mean()
print("Least score needed to be in top 20% of class:", val)


```

```python
# A p-value
# p-value is the probability of rejecting null hypothesis when the hypothesis
# is proven true

#Null Hypothesis: Common for students to score 68 marks.

#1.Calculcate zscore
zscore = (68-classscore.mean())/classscore.std()
print("Z-score", zscore)

#Calculate probability of getting this zscore and above
prob = 1- stats.norm.cdf(zscore)

print("The probability of having score of 68 and above:", prob)
#Probability is only 1.6% i.e p value is very small
# Hence null hypothesis can be rejected
```

```python
#Two tailed tests example
#mean score of math exam at national leve is 60 and std dev = 3
#Mean marks for class 53
# Null Hypothesis: Mean marks of the class are similar to national avg

#1.What is z score of class avg = 53 w.r.t to national avg
zscore = (53-60)/3.0
pvalue = stats.norm.cdf(zscore)
print("pvalue=",pvalue)
print(".98% is extremely small pvalue, hence rejecting the null hypothesis")
```

```python
# Confidence interval

#Random hieght data
#a= np.random.normal(loc = 150, scale = 10.0, size =[1,2,50]).round()
height_data = np.random.normal(loc = 183, scale = 10.0, size =100).round()
plt.hist(height_data, 30, normed= True) #30 = number of bins

print("Mean height", height_data.mean())
#For determining the confidence interval, we need standard error
std_error = height_data.std()/np.sqrt(len(height_data))
print("Standard error by calc", std_error)
print("Standard error by formula", stats.sem(height_data))

#95% confidence interfal  = 1.96 std deviations
#95% confidence limits = mean_height +/- sigma*std error

lower= height_data.mean()-1.96*stats.sem(height_data)
higher= height_data.mean()+1.96*stats.sem(height_data)

print("95% confidence interval: ", lower, higher)
```

```python
#Lets take sampple of 50 people and record their height and repeat proces
#30 times and find the average height

average_height =[]
for i in range(0,30):
    sample50 = np.random.normal(loc=183,scale=10, size= 50).round()
    average_height.append(sample50.mean())
plt.hist(average_height,20, normed = True)

#With sampling about 50 people 30 times, we have the average height
#distributed between 180 to 186
```

```python
#Lets take sampple of 100 people and record their height and repeat proces
#30 times and find the average height.
average_height =[]
for i in range(0,30):
    sample50 = np.random.normal(loc=183,scale=10, size= 1000).round()
    average_height.append(sample50.mean())
plt.hist(average_height,20, normed = True)

#With sampling about 50 people 30 times, we have the average height
#distributed between 182 and 183; ie. the range of average heights have
#narrowed down as we sample more.

#As the sample size increase the standard error of mean decreases and 
# hence confidence interval becomes narrower.
```

```python
#Correlation
mpg = [21.0, 21.0, 22.8, 21.4, 18.7, 18.1, 14.3, 24.4, 22.8,
19.2, 17.8, 16.4, 17.3, 15.2, 10.4, 10.4, 14.7, 32.4, 30.4,
33.9, 21.5, 15.5, 15.2, 13.3, 19.2, 27.3, 26.0, 30.4, 15.8,
19.7, 15.0, 21.4]

hp = [110, 110, 93, 110, 175, 105, 245, 62, 95, 123, 123, 180,
180, 180, 205, 215, 230, 66, 52, 65, 97, 150, 150, 245,
175, 66, 91, 113, 264, 175, 335, 109]

plt.scatter(mpg,hp)
plt.axis([5,40,0,400])
plt.xlabel("Miles per gallon")
plt.ylabel("Horsepower")

#Calcualte correlation coefficient
print("Pearson Correlation coefficient:", stats.pearsonr(mpg,hp)[0])
print("p value of Pearson Correlation coefficient:", 
      stats.pearsonr(mpg,hp)[1])

print("Spearman Correlation coefficient:", stats.spearmanr(mpg,hp)[0])
print("p value of spearman Correlation coefficient:", 
      stats.spearmanr(mpg,hp)[1])

#since the correlation value is -ve, there is inverse relationship between
# horsepower and mpg i.e as mpg increases the horsepower decreases.
```

```python
#Z-Test vs T-test
#T-Distribution is similar to z-distribution, centered around zero
# has bell shape, but its shorter and flatter around the center
# with fat tails
#Z-test compare population mean against a sample or compare the population
# mean of two distributions with sample size >30. Example: Comparing
#heights of men from different ethnicity groups.

#T-test is used to compare the population mean aginsta  sample or compare
#the population mean of two distributions with a sample size less <30
# and when st deviation is not knwn

class1_score = np.array([45.0, 40.0, 49.0, 52.0, 54.0, 64.0,
36.0, 41.0, 42.0, 34.0])

class2_score = np.array([75.0, 85.0, 53.0,70.0, 72.0, 93.0,
61.0, 65.0, 65.0, 72.0])

#Perform t-test to compare these populations
classttest = stats.ttest_ind(class1_score,class2_score)
print("T-Test statistic",classttest[0])
print("T-Test p-value ",classttest[1])

#NEED TO READ WHAT THIS EXACTLY MEANS
```
